```python
class minwoo:
    def __init__(self,hp,mp):
        self.hp=hp
        self.mp=mp
class JH(minwoo):
    def Fkiller(self):  # self == 인스턴스
        print("나는 칙칙이를 뿌린다.")
    def cal(self):
        print(self.hp+self.mp)
scv=JH(10,10)
scv.Fkiller()
scv.cal()
```

```python
#스레드 하기
from threading import Thread # 쓰레딩에서 쓰레드 클래스를 사용하겠다.
import time #타임 패키지를 사용한다.

def my_thread(val):
    for i in range(3):
        print("마이스레드")
        time.sleep(2)  # 2초간 스레드를 정지한다.
t1=Thread(target=my_thread,args=(1,))  # target은 스레드함수 이름, 두번째 매개변수를 튜플형태 
t1.start()
for i in range(3):
    print('메인함수에서 동작')
    time.sleep(1)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
n=100
xmin,xmax,ymin,ymax,zmin,zmax=0,20,0,20,0,50
cmin,cmax=0,2
xs=np.array([(xmax-xmin)*np.random.random_sample()+xmin for i in range(n)])
ys=np.array([(ymax-ymin)*np.random.random_sample()+ymin for i in range(n)])
zs=np.array([(zmax-zmin)*np.random.random_sample()+zmin for i in range(n)])
color=np.array([(cmax-cmin)*np.random.random_sample()+cmin for i in range(n)])

fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(xs,ys,zs,c=color,marker='o',s=15,cmap='Greens')
plt.show()
```

```python
#여러개 그래프 그리기
import matplotlib.pyplot as plt
import numpy as np

x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos( 2* np.pi * x1) *np.exp(-x1)
y2=np.cos( 2* np.pi * x2)
plt.subplot(2,1,1)    #2행 1열 첫번째에 그리겠다.
plt.plot(x1,y1,'o-')
plt.title('frequency graph')

plt.subplot(2,1,2)   #2행 2열 두번째꺼에 그리겠다.
plt.plot(x2,y2,'.-')
plt.title('cos graph')
plt.show()
```

```python
#박스 플롯
import matplotlib.pyplot as plt
import numpy as np

# 기본 스타일 설정하기
plt.style.use('default')
plt.rcParams['figure.figsize']=(4,3)  # 그림판을 4X3으로 만든다
plt.rcParams['font.size']=12

# 데이터준비하기
np.random.seed(0)
data_a=np.random.normal(0,2.0,1000)   # 0~2까지 천개 데이터
data_b=np.random.normal(-3.0,1.5,500)  # -3~1.5 500개 데이터 만들기
data_c=np.random.normal(1.2,1.5,1500)  # 1.2~ 1.5 1500개 만든다

# 그래프 그리기
fig, ax =plt.subplots()
ax.boxplot([data_a,data_b,data_c])
ax.set_ylim(-10.0,10.0)
ax.set_xlabel('DATA Type')
ax.set_ylabel('Value')
plt.show()
```

```python
# 객체지향 인터페이스
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,5)
fig, ax =plt.subplots(2,2,sharex=True,sharey=True,squeeze=True) # 2x2=4 개로 나눈다.
ax[0][0].plot(x,np.sqrt(x))
ax[0][1].plot(x,x,'bo')
ax[1][0].plot(x,-x+5)
ax[1][1].plot(x,np.sqrt(-x+5))
plt.show()
```

```python
#바이올린 플롯 그리기
import matplotlib.pyplot as plt
import numpy as np

#기본스타일 정하기
plt.style.use('default')
plt.rcParams['figure.figsize']=(4,3)
plt.rcParams['font.size']=12

#데이터 만들기
np.random.seed(0)  # 시간으로 랜덤값을 초기화 한다.
data_a=np.random.normal(0,2.0,1000)
data_b=np.random.normal(-10,3,500)
data_c=np.random.normal(1.3,1.7,1500)

print(len(data_a),len(data_b),len(data_c))
# 그래프 그리기
fig, ax=plt.subplots() # 객체로 서브플롯을 만든다.
ax.violinplot([data_a,data_b,data_c],positions=[2,3,4]) # 바이올린플롯을 그린다.
ax.set_ylim(-10.0,10.0)  #최대 최소값을 지정한다.
ax.set_xticks([1,2,3,4,5])
ax.set_xlabel('data type')
ax.set_ylabel('value')
plt.show()
```
